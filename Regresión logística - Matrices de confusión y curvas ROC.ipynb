#### PARTE 2. DUMMIEFICACIÓN DE LAS VARIABLES CATEGÓRICAS

```python
# librerías de scikit-learn para el modelo
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
# Se seleccionan las variables categóricas presentes en el dataset
categories = ['job', 'month', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'day_of_week', 'poutcome']

# copia de trabajo para la dummyficación
raw_dummydf = data.copy()

# dummyficación + joining para componer el dataset completo bruto
for category in categories:
    dummy = pd.get_dummies(data[category], prefix=category)
    raw_dummydf = raw_dummydf.join(dummy)

raw_dummydf.head()
```

```python
# estructura del bruto
raw_dummydf.shape
```

```python
# Se eliminan las features originales no dummyficadas
raw_dummydf = raw_dummydf.drop(categories, axis = 1)
raw_dummydf.columns.values
```

```python
# Se elimina la variable de salida para obtener
# Dataset dummyficado neto
data_dummyfied = raw_dummydf.drop(['y', 'y_bool'], axis = 1)
data_dummyfied.columns.values
```

```python
# Selección de la matriz bruta de features y del vector de salida
Y = data['y_bool'].values.ravel()
X = data_dummyfied
```

#### Selección de predictoras para el modelo

```python
# Número de variables a evaluar
n = 12
```

```python
# instanciación del objeto para el modelo
lr_model = LogisticRegression()
```

```python
# Ajuste del modelo por eliminación iterativa de predictoras: Recursive Feature Elimination RFE
'''
The goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features.
First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through any specific attribute or callable.
Then, the least important features are pruned from current set of features.
That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.
'''
rfe = RFE(lr_model, n)
rfe.fit(X , Y)
```

```python
# muestra de las predictoras escogidas
print(rfe.support_)
```

```python
# orden de importancia de todas las predictoras: 1: 12 escogidas; 2, 3, 4...: orden de importancia de las no escogidas
print(rfe.ranking_)
```

```python
# listado del ZIP de X, rfe.support_ y rfe.ranking_ para comprobar las predictoras que han quedado en el modelo
lrm_predictor_vars = zip(X, rfe.support_, rfe.ranking_)
list_lrm_vars = list(lrm_predictor_vars)
list_lrm_vars
```

```python
list_lrm_predictor_vars = []

for predictor in list_lrm_vars:
    
    if predictor[1] == True:
        print(predictor[0], '->', predictor[1])
        list_lrm_predictor_vars.append(predictor[0])

print('Variables predictoras:', list_lrm_predictor_vars)
```

```python
# Preparación para la predicción
X = data_dummyfied[list_lrm_predictor_vars]
Y = data['y_bool']
```

#### PARTE 5. VALIDACIÓN CRUZADA DEL MODELO

Los diferentes métodos de validación cruzada son utilizados amplia y generalmente para prevenir y evitar posibles problemas de overfitting en los modelos.

* **Método k-fold**: validación cruzada de k-elementos. El dataset original se divide en k particiones; de esas, una se utilizará como conjunto de testing; el resto, como conjunto de entrenamiento. Se repite el proceso k veces: cada una de las k particiones será utilizada una vez como conjunto de testing; así, al final, todas y cada una habrán actuado al menos una vez como conjunto de testing y de entrenamiento. Para cada iteración se medirá la eficacia del modelo de modo que después de todas las k iteraciones se tendrám k valores de eficacia y se podrá promediar la eficacia total del modelo, que será el promedio de los promedios obtenidos. 

#### Validación cruzada en scikit-learn

```python
# validación cruzada de scikit-learn
from sklearn.model_selection import cross_val_score
```

```python
# puntación de la validación cruzada para el dataset en uso. Scoring/objetivo: mejor eficacia; k=10
scores = cross_val_score(linear_model.LogisticRegression(), X, Y, scoring='accuracy', cv=10)
scores
```

```python
# Eficacia promedio de las eficacias de las particiones
scores.mean()
```

###### Sensibilidad y especificidad del modelo

* Sensibilidad (TPR) = TP/(TP+FN)

* Especificidad (TNR) = TN/(TN+FP)

```python
# conjuntos de entrenamiento y testing
X_train, X_testing, Y_train, Y_testing = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
Y_testing
```

```python
# creación del modelo
logit_model = linear_model.LogisticRegression()
logit_model.fit(X_train, Y_train)
```

```python
# array de probabilidades
probabilidades = logit_model.predict_proba(X_testing)
probabilidades
```

```python
prob = probabilidades[:,1]
prob_df = pd.DataFrame(prob)
threshold = 0.1
prob_df['prediction'] = np.where(prob_df>=threshold, 1, 0)
prob_df['actual'] = list(Y_testing)
prob_df
```

```python
prob
```

##### Tabla de confusión

```python
# tabla de confusión de 
confusion_matrix = pd.crosstab(prob_df['prediction'], prob_df['actual'])
confusion_matrix
```

```python
TN = confusion_matrix[0][0]
TP = confusion_matrix[1][1]
FN = confusion_matrix[1][0]
FP = confusion_matrix[0][1]
```

```python
sens = TP/(TP+FN)
sens
```

```python
specf_1 = 1-TN/(TN+FP)
specf_1 
```

```python
thresholds = [0.04, 0.07, 0.1, 0.12, 0.15, 0.18, 0.2, 0.25]

sensitivities = [1]
specificities = [1]

for thold in thresholds:

    prob_df['prediction'] = np.where(prob_df>=thold, 1, 0)
    prob_df['actual'] = list(Y_testing)
    prob_df.head()

    confusion_matrix = pd.crosstab(prob_df['prediction'], prob_df['actual'])

    TN = confusion_matrix[0][0]
    TP = confusion_matrix[1][1]
    FN = confusion_matrix[1][0]
    FP = confusion_matrix[0][1]

    sens = TP/(TP+FN)
    sensitivities.append(sens)
    
    specf_1 = 1-TN/(TN+FP)
    specificities.append(specf_1)

specificities.append(0)
sensitivities.append(0)
```

```python
sensitivities
```

```python
specificities
```

```python
%matplotlib inline
plt.plot(specificities, sensitivities, marker='o', linestyle="--", color='r')

x = [i*0.01 for i in range(100)]
y = [i*0.01 for i in range(100)]
plt.plot(x, y)
plt.xlabel('1-Especificidad')
plt.ylabel('Sensibilidad')
plt.title('Curva ROC')
```

#### Curva ROC con scikit-learn

```python
from sklearn import metrics
from ggplot import *
```

```python
specf_1, sensitivity, _ = metrics.roc_curve(Y_testing, prob)
```

```python
df = pd.DataFrame(
    {
        'spec': specf_1,
        'sens': sensitivity
    }
)
```

```python
df.head()
```

```python
# ploteo de la curva ROC con ggplot: estética aes + línea geométrica entre los puntos + recta de regresión 
ggplot(df, aes(x='spec',y='sens')) + geom_line() + geom_abline(linetype='dashed')
```

```python
# área bajo la curva ROC con sklearn: area under curve. Siendo auc = 1 el modelo perfecto.
auc = metrics.auc(specf_1, sensitivity)
auc
```

```python
# AUC con ggplot: estética aes transparencia 0.25
ggplot(df, aes(x='spec', y='sens')) + geom_area(alpha=0.25) + geom_line(aes(y='sens')) + ggtitle('Curva ROC y área bajo la curva AUC=%s' % str(auc)) + xlab('1-Especificidad') + ylab('Sensibilidad')
```
